```python
names = ['CO2', 'H2', 'CH3OH', 'H2O']
```

```python
from ase.build import molecule
from ase import Atoms
from ase.io import write
import os
import subprocess
```

```python
def create_dir(name) -> None:
    """Crea una carpeta dado un nombre"""
    try:
        os.mkdir(name)
        print(f"directory {name} is ready")
    except OSError as error:
        print(f"Error creating directory '{name}' : {error}")
```

```python
molecules: list[Atoms] = []
```

```python
for name in names:
    molecules.append(molecule(name, cell=[8, 8, 8]))
    create_dir(molecules[-1].get_chemical_formula())
    molecules[-1].write(f'{molecules[-1].get_chemical_formula()}/{molecules[-1].get_chemical_formula()}.xyz', format='xyz')
```

```python
os.system('sh run_xtb.sh')
```

```python
def convert_hartrees_to_ev(hartrees: float) -> float:
    return hartrees * 27.211
```

```python
cmd = ["sh", "read_energy.sh"]

e_producto = 0
e_reactivo = 0

for i in molecules:
    
    result = subprocess.run(cmd + [i.get_chemical_formula()], capture_output=True, text=True)

    if i.get_chemical_formula() == 'CO2':
        e_reactivo += float(result.stdout)
    elif i.get_chemical_formula() == 'H2':
        e_reactivo += 3 * float(result.stdout)
    elif i.get_chemical_formula() == 'CH4O':
        e_producto += float(result.stdout)
    else:
        e_producto += float(result.stdout)

    print(f"para {i.get_chemical_formula()}: {convert_hartrees_to_ev(float(result.stdout))} eV")
```

CO_2 + 3H_2 \rightarrow CH_3OH + H_2O
