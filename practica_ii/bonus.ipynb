```python
wave_number = []
ir_intensity = []

with open('./ir/naphthol/vibspectrum') as file:
    for line in file:
        line = line.split()
        if 'YES' in line:
            wave_number.append(float(line[2]))
            ir_intensity.append(float(line[3]))
```

```python
import matplotlib.pyplot as plt 
```

```python
plt.plot(wave_number, ir_intensity)
plt.xlabel('wavelength (cm**(-1))')
plt.ylabel('intensity (km*mol**(-1))')
plt.title('Teorical Infrared Spectrum (absdorbance)')
plt.show()
```
