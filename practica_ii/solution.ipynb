- Cada átomo tiene 3 coordenadas: $x$, $y$ y $z$
- Si tiene $N$ átomos, tiene en total $3N$ coordenadas independientes
Por eso, la matriz hessiana debe contener todas las combinaciones posibles de derivadas entre esas coordenas: $3N \times 3N$

Ejemplo sencillo:
- Si en una molécula se tienen 2 átomos ($N = 2$)
  - Hay 6 coordenadas: $x_1, y_1, z_1, x_2, y_2, z_2$
  - Entonces la Hessiana será de tamaño $6 \times 6$

## Punto 2.b

Claro, resolvamos todo el punto que planteas de forma completa:

---

🔹 **i) ¿Qué es el número de onda?**

El **número de onda** es una cantidad que indica **cuántas longitudes de onda caben en una unidad de distancia**.  
En espectroscopía infrarroja (IR) y en vibraciones moleculares, el número de onda se define como:

$$
\tilde{\nu} = \frac{1}{\lambda}
$$

donde:
- $( \tilde{\nu} $ es el número de onda,
- $\lambda $ es la longitud de onda (en centímetros, generalmente).

El número de onda representa la frecuencia espacial de la onda: **cuántos ciclos hay por centímetro**.

En este contexto, **se mide en unidades de centímetros inversos** $(\text{cm}^{-1})$.

---

🔹 **ii) ¿Qué relación tiene el número de onda con la energía de cada vibración?**

La **energía** $ E $ de una vibración está relacionada directamente con el número de onda $ \tilde{\nu} $ por medio de la fórmula:

$$
E = h c \tilde{\nu}
$$

donde:
- $ h $ es la constante de Planck ($6.626 \times 10^{-34}\ \text{J}\cdot\text{s}$),
- $ c $ es la velocidad de la luz ($3.00 \times 10^8\ \text{m/s}$),
- $ \tilde{\nu} $ es el número de onda en $\text{cm}^{-1}$.

**Interpretación**:  
- A mayor número de onda, **mayor energía** de la vibración.
- Por eso, en espectros IR, los picos en regiones de mayor $ \tilde{\nu} $ (por ejemplo, 3000 $\text{cm}^{-1}$) corresponden a vibraciones más energéticas, como estiramientos de enlaces simples ligeros (como C-H).

---

🔹 **iii) ¿Cómo entender que se exprese en unidades de cm$^{-1}$?**

Cuando se mide en $\text{cm}^{-1}$, el número de onda te dice **cuántas ondas caben en un centímetro**.  
- Si $ \tilde{\nu} = 1000\ \text{cm}^{-1} $, significa que hay 1000 ciclos de la onda en cada centímetro.
- Esta unidad es muy práctica en espectroscopía porque las longitudes de onda del infrarrojo son del orden de micrómetros ($\mu m$) o fracciones de centímetro, y usar $\text{cm}^{-1}$ da números cómodos (como 500, 1000, 3000) en vez de valores muy pequeños o muy grandes.

## Punto 3.b
### Orbitales atómicos

Los orbitales moleculares se forman cuando dos átomos se acercan y sus orbitales atómicos se combinan.
Est acombinación puede ser de dos tipos principales:

#### Orbitales moleculares enlazantes
- Se forman cuando las ondas de los orbitales atómicos se suman de forma constructiva (sus fases coinciden).
- **Resultado**: aumenta la densidad electrónica entre los dos núcleos.
- Esto **estabiliza** la molécula por que los núcleos comparten más carga electrónica entre ellos, formando un **enlace fuerte**.
- Energéticamente, el orbital enlazante **tiene menos energía** que loos orbitales originales.

#### Orbitales moleculares antienlazantes
- Se forman cuando las ondas de los orbitales atómicos **se combinan de manera destructiva** (sus fases son opuestas).
- **Resultado**: se crea una**zona de baja densidad electrónica** entre los núcleos para mantenerlos unidos; **genera repulsión**.

### HOMO
HOMO (Highest Occupied Molecular Orbital) es simplemente:
> **El orbital molecular de más alta energía que todavía tiene electrones ocuándolo.**

En otras palabras:
- Es **el últimmo orbital lleno** cuando acomodas los electrones en los orbitales moleculares, siguiendo las reglas de ocupación (como el principio de Aufbau).
- Es muy importante porque:
  - **Puede donar electrones** fácilmente (por ejemplo, en reacciones químicas).
  - Determinar en parte la **reactividad** de la molécula.

### LUMO
El **LUMO** (Lowest Unoccupied Molecular Orbital) es:
> **El primer orbital molecualr que está vacío, justo encima del HOMO.**

En otras plabras:
- es el orbital de menor energía que aún no tiene electrones.
- Es importante porque:
  - Puede aceptar electrones fácilmente.
  - También determian parte de la reactividad de la molécula, especialmente en reacciones donde se reciben electrones (por ejemplo, en reacciones de adición o transferencia de carga).
